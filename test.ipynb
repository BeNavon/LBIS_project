```python
import pandas as pd
import matplotlib.pyplot as plt
import glob
import os

# Get all CSV files matching the pattern
csv_files = glob.glob('data/*/*/treadmill/gcRight/*.csv')
csv_file = glob.glob('data/*/*/treadmill/gcLeft/*.csv')

# Create a figure outside the loop to plot all data on same graph
plt.figure(figsize=(12, 6))

# Plot each CSV file
for file in csv_files:
    # Read CSV file
    df = pd.read_csv(file)
    
    # Get subject and condition from path
    path_parts = os.path.dirname(file).split(os.sep)
    subject = path_parts[-4]
    condition = path_parts[-3]
    
    # Plot each column with its header name in the legend
    for column in df.columns:
        plt.plot(df[column],'.', label=f'{column} ({subject}-{condition})')

    for file in csv_file:
        # Read CSV file
        df = pd.read_csv(file)
        
        # Get subject and condition from path
        path_parts = os.path.dirname(file).split(os.sep)
        subject = path_parts[-4]
        condition = path_parts[-3]
        
        # Plot each column with its header name in the legend
        for column in df.columns:
            plt.plot(df[column],'.', label=f'{column} ({subject}-{condition})')
        
        break
    break

plt.title('Gait Cycle - Right Leg', fontsize=14, pad=20)
plt.xlabel('Percentage of Gait Cycle (%)', fontsize=12)
plt.ylabel('Ground Contact (1 = Contact, 0 = No Contact)', fontsize=12)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.xlim(5000,5500)
plt.show()
```
